```python
sns.countplot(data=df,x='test_result')
```

we can see that there are total 2000 people fail the test and 3000 people pass the test

```python
col_list=df.columns.to_list()
col_list.pop(2) #poping out test_result
col_list
```

```python
for col in col_list:
    plt.figure(figsize=(10,8))
    sns.boxplot(data=df, y=col,x='test_result')
    plt.show()
```

- from 1st fig we clearly see that people with more age tends to fail in test whether people with low age are passing the test
- from 2nd figure we can see the people with high physical score are passing and people with low physical score are failing

```python
#lets check the relationship of both indp varible with hue dep varible
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='age',y='physical_score',hue='test_result',alpha=0.5)
```

- people having lower age and high score will pass the test

```python
plt.figure(figsize=(15,15))
sns.pairplot(df,hue='test_result')
```
